```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
review_cnt = soup.find('span', class_= 'index__tabCount__3oL0k')
# index__tabCount__3oL0k class containing the number of reviews 
```

```python
# click action required to load all reviews
from selenium.webdriver.common.action_chains import ActionChains
j =10
for i in range(0, 600): 
    try:
        driver.find_element('xpath','//*[@id="reviews-read-more-cta"]').click()  
        print(j)
        print(f'{i}th iteration running')
        print('***************************')
        # loading period
        time.sleep(2)
    except: 
        break
    j =j+10
```
